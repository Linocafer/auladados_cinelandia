```python
import pandas as pd 
import numpy as np
```

```python
df_anp = pd.read_csv('ANP2025.csv', sep=';')    
df_anp
```

```python
# Ver as primeiras linhas
print(df_anp.head())

# Informações gerais: colunas, tipos e valores nulos
print(df_anp.info())

# Estatísticas descritivas das colunas numéricas
print(df_anp.describe())
```

```python
# Filtrar apenas linhas do Rio de Janeiro
df_anp_rj = df_anp.loc[df_anp['Estado - Sigla'] == 'RJ']

# Mostrar as primeiras linhas do novo DataFrame
df_anp_rj.head(5)
```

```python
print(df_anp_rj.dtypes)
```

```python
# Substituir vírgula por ponto (se houver) e remover espaços
df_anp_rj['Valor de Venda'] = df_anp_rj['Valor de Venda'].str.replace(',', '.').str.strip()

# Converter para float
df_anp_rj['Valor de Venda'] = df_anp_rj['Valor de Venda'].astype(float)
```

```python
media_munic_rj = df_anp_rj.groupby('Municipio')['Valor de Venda'].mean().reset_index()
media_munic_rj = media_munic_rj.sort_values(by='Valor de Venda', ascending=False)
media_munic_rj
```

```python
array_valor_venda = np.array(media_munic_rj['Valor de Venda'])
array_valor_venda
```

```python
# Calcular média e mediana
media = np.mean(array_valor_venda)
mediana = np.median(array_valor_venda)

# Supondo que array_valor_venda seja o array NumPy
q1 = np.quantile(array_valor_venda, 0.25)  # 25% (1º quartil)
q2 = np.quantile(array_valor_venda, 0.50)  # 50% (mediana)
q3 = np.quantile(array_valor_venda, 0.75)  # 75% (3º quartil)

# Mostrar Resultados
print(f'A media é {media}') 
print(f'A mediana é {mediana}') 
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2, Mediana): {q2}')
print(f'Terceiro quartil (Q3): {q3}')



```

```python
munic_acima_q3 = media_munic_rj.loc[media_munic_rj['Valor de Venda'] > q3]
munic_acima_q3
```
